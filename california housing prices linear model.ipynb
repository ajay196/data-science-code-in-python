```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
df=pd.read_csv("/home/ajay/data science/data-master/california_housing_train.csv")
df.head()
```

```python
y_train=df.median_house_value
y_train.head()
```

```python
x_train=df.drop(['median_house_value'],axis=1)
x_train.head()
```

```python
df2=pd.read_csv("/home/ajay/data science/data-master/california_housing_test.csv")
df2.head()
```

```python
y_test=df2.median_house_value
y_test.head()
```

```python
x_test=df2.drop(['median_house_value'],axis=1)
x_test.head()
```

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error,r2_score
```

```python
model=linear_model.LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
print("modle coefficient :",model.coef_)
print("model intercept:",model.intercept_)
```

```python
print("model mean square error is %.4f"%mean_squared_error(y_test,y_pred))
print("model coefficient of determination is :%.4f"%r2_score(y_test,y_pred))
```
